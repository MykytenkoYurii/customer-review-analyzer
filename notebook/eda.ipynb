```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
with open("../data/processed.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)
df.head()
```

```python
sentiment_counts = df["sentiment"].value_counts()

plt.figure(figsize=(6,4))
sentiment_counts.plot(kind="bar", color="skyblue")
plt.title("üìä –†–æ–∑–ø–æ–¥—ñ–ª –≤—ñ–¥–≥—É–∫—ñ–≤ –∑–∞ —Ç–æ–Ω–∞–ª—å–Ω—ñ—Å—Ç—é")
plt.xlabel("–¢–æ–Ω–∞–ª—å–Ω—ñ—Å—Ç—å")
plt.ylabel("–ö—ñ–ª—å–∫—ñ—Å—Ç—å")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
all_keywords = sum(df["keywords"], [])
keyword_counts = Counter(all_keywords).most_common(10)

keywords, counts = zip(*keyword_counts)

plt.figure(figsize=(8,5))
plt.barh(keywords[::-1], counts[::-1], color="lightgreen")
plt.title("üîë –¢–æ–ø-10 –∫–ª—é—á–æ–≤–∏—Ö —Å–ª—ñ–≤")
plt.xlabel("–ß–∞—Å—Ç–æ—Ç–∞")
plt.tight_layout()
plt.show()
```
